```python
import os 
import librosa
import pandas as pd
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
```

```python
GENRE_LIST=['metal','hiphop','reggae','rock','classical','jazz','blues','country','pop','disco']
print("Genre list ",GENRE_LIST)
```

```python
#Todo modify method to return dataframe
def CreateDataSet(folder):
    spectralcentroids=[]
    samplerate=[]
    label=[]
    path=[]
    for directory in folder:
        columns=['data', 'samplerate', 'label', 'path']
        df = pd.DataFrame(columns=columns)
        for filename in os.listdir(directory):
            music_name=os.path.join(directory,filename)
            try:
                X, sample_rate = librosa.load(music_name, res_type='kaiser_fast')
                spectralcentroid = librosa.feature.spectral_contrast(y=X, sr=sample_rate , hop_length=512)
                centroid = np.mean(np.array(spectralcentroid))
                std = np.std(np.array(spectralcentroid))
                spectralcentroids.append((centroid,std))
                samplerate.append(sample_rate)
                path.append(music_name)
                label.append(directory)
            except Exception as e:
                print("Error encountered while parsing file: ", e)
                return None
    return np.array(spectralcentroids), np.array(samplerate), np.array(label), np.array(path)
```

```python
centroid, samplerate, label, path = CreateDataSet(GENRE_LIST)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(label)

onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D, Conv2D
from keras.optimizers import Adam
from keras.utils import np_utils
from sklearn import metrics
model = Sequential()
model.add(Dense(32, input_shape=(2,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
model.compile(loss='mean_squared_error', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(centroid, onehot_encoded, batch_size=32, epochs=500)
```
