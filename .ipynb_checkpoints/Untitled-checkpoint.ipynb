```python
import os 
import librosa
import pandas as pd
```

```python
print(os.listdir())
```

```python
#create a data set with labels
```

```python
def CreateDataSet(path):
    df = pd.DataFrame(columns=['data', 'samplerate', 'label'])
    for filename in os.listdir(path):
        music_name=os.path.join(path,filename)
        os.system("sox " + str(music_name) + " " + str(music_name[:-3]) + ".wav")
        os.remove(music_name)
        music_name = music_name[:-3] + '.wav'
        try:
            X, sample_rate = librosa.load(music_name, res_type='kaiser_fast')
            df.append(pd.Series([X,sample_rate,path],index=['data', 'samplerate', 'label']),ignore_index=True)
        except Exception as e:
            print("Error encountered while parsing file: ", music_name)
    return df
```

```python
df = pd.DataFrame(columns=['data', 'samplerate', 'label'])
for name in os.listdir():
    if os.path.isdir(name):
        tempdataframe=CreateDataSet(name)
        df = df.append(tempdataframe)
```
