```python
import os 
import librosa
import pandas as pd
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
```

```python
GENRE_LIST=['metal','hiphop','reggae','rock','classical','jazz','blues','country','pop','disco']
print("Genre list ",GENRE_LIST)
```

```python
#checking out a random audio file 
def plotdata(name):
    Data , SampeRate = librosa.load(name)
    plt.figure(figsize=(12, 4))
    librosa.display.waveplot(Data, sr=SampeRate)
    plt.title(name)
```

```python
name = ["blues/blues.00000.wav",'metal/metal.00000.wav','hiphop/hiphop.00000.wav','reggae/reggae.00000.wav','rock/rock.00000.wav','classical/classical.00000.wav']
for n in name:
    plotdata(n)
```

```python
spectralcentroids=[]
samplerate=[]
label=[]
path=[]
```

```python
#Todo modify method to return dataframe
def CreateDataSet(folder):
    columns=['data', 'samplerate', 'label', 'path']
    df = pd.DataFrame(columns=columns)
    for filename in os.listdir(folder):
        music_name=os.path.join(folder,filename)
        try:
            X, sample_rate = librosa.load(music_name, res_type='kaiser_fast')
            samplerate.append(sample_rate)
            label.append(folder)
            path.append(music_name)
            spectralcentroid = np.mean(librosa.feature.spectral_centroid(y=X,sr=sample_rate).T, axis=0)
            print(spectralcentroid.shape)
            spectralcentroids.append(spectralcentroid)
        except Exception as e:
            print("Error encountered while parsing file: ", e)
            return None
    return df
```

```python
for folder in GENRE_LIST:
    print(folder)
    CreateDataSet(folder)
```

```python
ndata = np.array(spectralcentroids)
y = np.array(label)
```

```python
ndata.shape
```

```python
lb = LabelEncoder()

y = np_utils.to_categorical(lb.fit_transform(y))
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D, Conv2D
from keras.optimizers import Adam
from keras.utils import np_utils
from sklearn import metrics 

num_labels = y.shape[1]

model = Sequential()

model.add(Dense(13, input_dim=800, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
```

```python
model.compile(optimizer='adam', 
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
X_train, X_val, y_train, y_val = train_test_split(ndata,y,test_size=.2,random_state=1)
```

```python
model.fit(X_train, y_train, batch_size=32, epochs=500,validation_data=(X_val,y_val))
```
